```python
from flask import Flask, request, render_template, send_file, jsonify
import torch
from torch import autocast
from diffusers import DiffusionPipeline
from io import BytesIO
from pyngrok import ngrok
import base64

app = Flask(__name__)

device = "cuda"
model_id = "sd-dreambooth-library/tattoo-design"
pipe = DiffusionPipeline.from_pretrained("sd-dreambooth-library/tattoo-design")
pipe.to(device)

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/getprediction', methods=['POST'])
def getprediction():
    prompt = request.form['prompt']
    full_prompt = "line art sks tattoo design " + prompt
    images = []

    with autocast(device):
        for _ in range(3):  # Generate three images
            try:
                print(f"Generating image with prompt: {full_prompt}")
                image = pipe(prompt=full_prompt, guidance_scale=8.5).images[0]
            except Exception as e:
                print(f"Error generating image: {e}")
                return jsonify(error=str(e)), 500

            buffer = BytesIO()
            image.save(buffer, format="PNG")
            buffer.seek(0)
            imgstr = base64.b64encode(buffer.getvalue()).decode('utf-8')
            images.append(imgstr)

    return jsonify(images=images)

if __name__ == "__main__":
    ngrok.set_auth_token("2gaClo7HeTvEm8BbyBENgbTyULK_7HxLMvc1g9pU42N1PSfGT")  # Replace 'YOUR_AUTHTOKEN' with your actual ngrok auth token
    ngrok_tunnel = ngrok.connect(5000)
    print('Public URL:', ngrok_tunnel.public_url)
    app.run(port=5000)
```
